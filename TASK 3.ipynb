```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
df = pd.read_csv('Country-data.csv')

# Set the style for better visualizations
plt.style.use('seaborn-v0_8')  # Changed from 'seaborn' to 'seaborn-v0_8' which is a valid style name
# Alternative styles: 'ggplot', 'fivethirtyeight', 'seaborn-whitegrid', etc.

# 1. Bar Chart: Top 10 countries by income
plt.figure(figsize=(12, 6))
top_10_income = df.nlargest(10, 'income')
plt.bar(top_10_income['country'], top_10_income['income'])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Countries by Income')
plt.xlabel('Country')
plt.ylabel('Income')
plt.tight_layout()
plt.show()

# 2. Histogram: Distribution of Life Expectancy
plt.figure(figsize=(10, 6))
plt.hist(df['life_expec'], bins=30, edgecolor='black')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy (years)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Scatter Plot: Income vs Life Expectancy
plt.figure(figsize=(10, 6))
plt.scatter(df['income'], df['life_expec'], alpha=0.5)
plt.title('Income vs Life Expectancy')
plt.xlabel('Income')
plt.ylabel('Life Expectancy (years)')
plt.tight_layout()
plt.show()

# 4. Box Plot: Child Mortality Distribution by Health Spending
df['health_category'] = pd.qcut(df['health'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
plt.figure(figsize=(10, 6))
sns.boxplot(x='health_category', y='child_mort', data=df)
plt.title('Child Mortality Distribution by Health Spending Categories')
plt.xlabel('Health Spending Category')
plt.ylabel('Child Mortality Rate')
plt.tight_layout()
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(12, 8))
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numeric Variables')
plt.tight_layout()
plt.show()
```
